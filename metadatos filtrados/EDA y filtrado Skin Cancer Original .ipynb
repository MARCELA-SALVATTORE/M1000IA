```python
# URL del archivo CSV en GitHub
url = 'https://raw.githubusercontent.com/mvegetti/IAcourse/main/HAM10000_metadata.csv'

# Leer el archivo CSV desde la URL
df = pd.read_csv(url)

# Primeras filas del dataset
print("\nPrimeras 5 filas del dataset:")
print(df.head())

# Nombres de las columnas
print("\nNombres de las columnas:")
print(df.columns)

# Verificar los tipos de datos de cada columna
print("\nTipos de datos por columna:")
print(df.dtypes)


# Dimensiones del dataset (filas, columnas)
print("\nDimensiones del dataset (filas, columnas):")
print(df.shape)

# Información general del dataset
print("\nInformación general del dataset:")
print(df.info())

# Estadísticas descriptivas de las columnas numéricas
print("\nEstadísticas descriptivas:")
print(df.describe())

# Estadísticas descriptivas para todas las columnas (numéricas y categóricas)
print("\nEstadísticas descriptivas para todas las columnas:")
print(df.describe(include='all'))


```

```python
# Contar valores únicos por cada columna
print("\nValores únicos por columna:")
valores_unicos = df.nunique()
print(valores_unicos)

# Graficar el conteo de valores únicos en columnas categóricas y mostrar valores por escrito
for col in df.select_dtypes(include=['object']).columns:
    print(f"\nColumna: {col}")
    print(f"Valores únicos: {df[col].nunique()}")
    print(f"Conteo de cada valor en '{col}':")
    print(df[col].value_counts())

    # Graficar el conteo de valores
    plt.figure(figsize=(10, 5))
    sns.countplot(y=col, data=df, order=df[col].value_counts().index, palette='Set2')
    plt.title(f"Conteo de valores en {col}")
    plt.xlabel("Frecuencia")
    plt.ylabel(col)
    plt.show()
```

```python
# Mostrar las etiquetas (nombres de las columnas) del DataFrame
print("Etiquetas (nombres de las columnas) del dataset:")
print(df.columns.tolist())

# Identificar las columnas numéricas
columnas_numericas = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
cantidad_numericas = len(columnas_numericas)
print("\nCantidad de columnas numéricas:", cantidad_numericas)
print("Columnas numéricas en el dataset:")
print(columnas_numericas)

# Identificar las columnas no numéricas
columnas_no_numericas = df.select_dtypes(exclude=['float64', 'int64']).columns.tolist()
cantidad_no_numericas = len(columnas_no_numericas)
print("\nCantidad de columnas no numéricas:", cantidad_no_numericas)
print("Columnas no numéricas en el dataset:")
print(columnas_no_numericas)

# Calcular la correlación solo entre columnas numéricas
if columnas_numericas:
    correlacion = df[columnas_numericas].corr()
    print("\nMatriz de correlación entre las columnas numéricas:")
    print(correlacion)

    # Mostrar el mapa de calor de correlación
    plt.figure(figsize=(10, 6))
    sns.heatmap(correlacion, annot=True, cmap='coolwarm', fmt=".2f")
    plt.title("Mapa de calor de la correlación")
    plt.show()
else:
    print("\nNo hay columnas numéricas para calcular la correlación.")

# Guardar los resultados en un archivo de texto
with open('resultados_columnas.txt', 'w') as f:
    # Escribir resultados en el archivo
    f.write("Etiquetas (nombres de las columnas) del dataset:\n")
    f.write(str(df.columns.tolist()) + "\n")

    f.write("\nCantidad de columnas numéricas: {}\n".format(cantidad_numericas))
    f.write("Columnas numéricas en el dataset:\n")
    f.write(str(columnas_numericas) + "\n")

    f.write("\nCantidad de columnas no numéricas: {}\n".format(cantidad_no_numericas))
    f.write("Columnas no numéricas en el dataset:\n")
    f.write(str(columnas_no_numericas) + "\n")

    if columnas_numericas:
        f.write("\nMatriz de correlación entre las columnas numéricas:\n")
        f.write(str(correlacion) + "\n")

```

### Filtrado de valores nulos y unknown

```python
# Eliminar filas con valores nulos
df_filtrado = df.dropna()
df_filtrado.info()

# Eliminamos filas que tienen valores nulos en 'age' y valor 'unknown' en localization o sex
# y filas que tienen 'unknown' en 'sex' o 'localization'
df_filtrado = df_filtrado[(df_filtrado['sex'] != 'unknown') & (df_filtrado['localization'] != 'unknown')]  # Filtra filas con 'unknown'


```

###Filtrado de etiquetas que queremos
Estas son las etiquetas totales que tiene el dataset:

Etiqueta cantidad	Descripción 		 	          Tipo
akiec 	327	akiec  (Queratosis actínica y carcinoma intraepitelial /
enfermedad de Bowen)  Pre-Maligno
 bcc 	509	bcc (Carcinoma Basocelular)	Maligno
 bkl 	1076	bkl (Queratosis Benigna) Benigno
df 	115	df (Dermatofibroma)	Benigno
mel	1101	mel (Melanoma)	Maligno
nv 	6491	nv (Nevo o Lunar)	Benigno
vasc	142	asc (Lesiones Vasculares)	Benigno



De este dataset se descartaran las imagenes para las etiquetas: akiec - df - vasc


```python
# Valores que quieres eliminar
valores_a_eliminar = ['akiec', 'df', 'vasc']

# Filtrar el DataFrame para eliminar las filas con esos valores en 'dx'
df_filtrado = df_filtrado[~df_filtrado['dx'].isin(valores_a_eliminar)]

# Mostrar el DataFrame resultante
print(df_filtrado)
```

#eliminación de imagenes que correspondan a nevo

```python
# Filtrar los registros que tienen dx igual a 'NV'
filas_a_eliminar = df_filtrado[df_filtrado['dx'].isin(['nv'])]

# Seleccionar aleatoriamente 3500 registros de imagenes NV  para eliminar
if len(filas_a_eliminar) > 3954:
    filas_a_eliminar = filas_a_eliminar.sample(n=4100, random_state=1)  # random_state para reproducibilidad

# Eliminar las filas seleccionadas del DataFrame original
df_filtrado = df_filtrado.drop(filas_a_eliminar.index)

# Mostrar el DataFrame resultante
df_filtrado.shape
```

###Agregado de columnas

*   etiqueta maligno/benigno
*   identificación del dataset



```python
# Definir una función para clasificar
def clasificar(dx):
    if dx in ['nv', 'bkl']:
        return 'Lesión Benigna'
    elif dx in ['bcc', 'mel']:
        return 'Lesión Maligna'
    else:
        return None  # Puedes ajustar esto si deseas un valor por defecto

# Agregar la columna 'Clasificacion'
df_filtrado['Clasificacion'] = df_filtrado['dx'].apply(clasificar)

# Agregar la columna 'Extra'
df_filtrado['Dataset'] = 'Original'
df_filtrado.head()
```

### guardado del dataset en disco
