### Filtrado de columnas que  interesan para la unión
Como se observa de la información anterior, este dataset tiene más columnas de las que necesitamos y algunas de las que necesitamos tienen otros nombres. Debemos acomodar el dataset para que tenga las mismas columnas que el dataset que tenemos que completar (lesion_id, image_id, dx, dx_type,sex,age, localization). Para ello tenemos que eliminar columnas innecesarias y renombrar otras.

DatasetOriginal -> EsteDataset

lesion_id --> lesion_id

image_id --> img_id

dx --> diagnostic

dx_type --> "no existes"

sex --> gender

age --> age

localization --> region




```python
#Eliminar columnas innecesarias
# Seleccionar solo las columnas necesarias
columnas_necesarias = ['lesion_id', 'img_id', 'diagnostic', 'age', 'gender', 'region']
df_filtrado = df[columnas_necesarias]

# Mostrar las primeras filas del nuevo dataframe
df_filtrado.head()
df_filtrado.info()
```

### renombrar columnas


```python
# Renombrar las columnas
df_filtrado.rename(columns={
    'img_id': 'image_id',
    'diagnostic': 'dx',
    'gender': 'sex',
    'region': 'localization'
}, inplace=True)

# Mostrar el DataFrame con las columnas renombradas
print(df_filtrado)
```

### Filtrado de valores nulos y unknown

```python

# Eliminar filas con valores nulos
df_filtrado = df_filtrado.dropna()
df_filtrado.info()

```

###Filtrado de etiquetas que queremos
Estas son las etiquetas totales que tiene el dataset:

Etiqueta cantidad	Descripción 		 	          Tipo

ACK	       283		ACK (Queratosis actinica)	  Benigno

BCC     	 845		BCC (Carcinoma Basocelular)	    Maligno

MEL	        52		MEL (Melanoma)			            Maligno

NEV	        75		NEV (Nevo o Lunar)		        Benigno

SCC	       192		SCC (Carcinoma Espinocelular-Enf Bowen)	  Maligno

SEK	        47		SEK (Queratosis Seborreica)	    Benigno

De este dataset se descartaran las imagenes para las etiquetas NEV,SCC y ACK


```python
# Valores que quieres eliminar
valores_a_eliminar = ['ACK', 'NEV', 'SCC']

# Filtrar el DataFrame para eliminar las filas con esos valores en 'dx'
df_filtrado = df_filtrado[~df_filtrado['dx'].isin(valores_a_eliminar)]

# Mostrar el DataFrame resultante
print(df_filtrado)
```

###Agregado de columnas

*   etiqueta maligno/benigno
*   identificación del dataset



```python
# Definir una función para clasificar
def clasificar(dx):
    if dx in ['NEV', 'SEK']:
        return 'Lesión Benigna'
    elif dx in ['BCC', 'MEL']:
        return 'Lesión Maligna'
    else:
        return None  # Puedes ajustar esto si deseas un valor por defecto

# Agregar la columna 'Clasificacion'
df_filtrado['Clasificacion'] = df_filtrado['dx'].apply(clasificar)

# Agregar la columna 'Extra'
df_filtrado['Dataset'] = 'Extra'
df_filtrado.head()
```

### guardado del dataset en disco
